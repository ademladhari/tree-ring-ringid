# RingID / Tree-Ring Generation in Google Colab

Clone the repository, install requirements, and run Tree-Ring generation.

```python
# 1) Clone the repository
from pathlib import Path

REPO_URL = "https://github.com/ademladhari/tree-ring-ringid"
REPO_DIR = Path("tree-ring-ringid")

if not REPO_DIR.exists():
    !git clone {REPO_URL}
else:
    print(f"Repository already exists at {REPO_DIR}")

%cd tree-ring-ringid
```

```python
# 2) Install requirements
!python -m pip install -U pip
!pip install -r requirements.txt
!pip install -e .
```

```python
# 3) Run generation
PROMPT = "a photo of a red bicycle parked beside a brick wall"
OUT_DIR = "batch_run"

!ringid-generate --prompt "{PROMPT}" --count 40 --out-dir "{OUT_DIR}" --profile tree_ring --model-id runwayml/stable-diffusion-v1-5 --steps 50 --height 512 --width 512 --guidance-scale 7.5 --seed 42

!ls -lah "{OUT_DIR}"
```
